```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from Modules.brute_force_bond_finder import brute_force_bond_finder
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'grid'])
```

This notebook will compare the size of the full partition lattice $\Pi_n$ - which is equal to the Bell number $B_n$ for an $n$ element set - with the size of the bond lattice for certain key graphs we have analysed. We know that the bond lattice of a complete graph with $n$ vertices, $K_n$, is isomorphic to $\Pi_n$, hence we will use the number of bonds found by our algorithm for the complete graphs to show the scaling of the partition lattice.

```python
interval = np.arange(0, 9)

lattice_sizes = np.zeros((4, 9))
for n in interval:
    path = nx.path_graph(n)
    cycle = nx.cycle_graph(n)
    comp = nx.complete_graph(n)
    wheel = nx.wheel_graph(n)
    lattice_sizes[0, n] = len(brute_force_bond_finder(comp))
    lattice_sizes[1, n] = len(brute_force_bond_finder(wheel))
    lattice_sizes[2, n] = len(brute_force_bond_finder(cycle))
    lattice_sizes[3, n] = len(brute_force_bond_finder(path))
```

```python
print(lattice_sizes[3, -1])
```

```python
legend = {0 : "Complete", 1 : "Wheel", 2 : "Cycle", 3 : "Path"}

plt.figure(figsize=(12,8))

for i in range(4):
    plt.plot(lattice_sizes[i, :], label = legend[i], marker = "x")

plt.xlabel("Number of Variables (d)", fontsize=25)
plt.ylabel("Lattice Size", fontsize=25)
plt.grid(alpha=0.5)
plt.title("Plot of Lattice Size vs Number of Variables", fontsize=35)
plt.legend(fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig(r"C:\Users\morit\Thesis_Repo\Plots\lattice_size_plot.png")
plt.show()
```
